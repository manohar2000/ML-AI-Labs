```python
def predict(X,weights):
    return np.dot(X,weights)    
```

```python
def error(y_pred,y_actual):
    mse = (1/2*y_actual[0].shape) * np.sum((y_pred-y_actual)**2)
    return mse
```

```python
def gradient(X,y_true,y_pred):
    gradient = []
    
    gradient[0][0] = (1/X[0].shape) * np.sum(np.dot(X[:,0],(y_pred-y_true)))
    gradient[0][1] = (1/X[1].shape) * np.sum(np.dot(X[:,0],(y_pred-y_true)))
    
    return gradient
    
```

```python
def compiling(X,y,learning_rate = 0.0001, iterations = 1000):
    
    weights = np.random.randn((2,1))   
    losses = []
    
    y_true = y.reshape(-1,-1)
    
    for i in range(iterations):
        y_pred = predict(X,weights)
        
        mse = error(y_true,y_pred) 
        
        losses.append(mse)
        
        gradient = gradient(X,y_pred,y_true)
        
        weights[0][0] = weights[0][0] - learning_rate*gradient[0][0]
        weights[1][0] = weights[1][0] - learning_rate*gradient[1][0]
        
        
        
        
        
    return weights,losses
```
