```python
# Import the required libraries
# Use terminal commands like "pip install numpy" to install packages
import os
import numpy as np
import pandas as pd
#import pydicom
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import PIL
```

```python
img = mpimg.imread('data/rgb_images/1.jpeg')
print(img)
```

```python
img.shape
```

```python
imgplot = plt.imshow(img)
```

```python
img_shortened = img[400:1000,:,:]
img_shortened.shape
```

```python
image_shortened = plt.imshow(img_shortened)
```

```python
rgb_weights = [0.2989, 0.5870, 0.1140]
grayscale_image = np.dot(img_shortened[...,:3], rgb_weights)

print(grayscale_image)
```

```python
plt.imshow(grayscale_image, cmap=plt.get_cmap("gray"))
```

## Reading DICOM Images

```python
import matplotlib.pyplot as plt
import pydicom
import numpy as np

filename = pydicom.read_file('data/dicom_images/0002.DCM')

print(filename)
```

```python
print(filename.pixel_array.shape)
print(filename.pixel_array)
```

```python
arr = filename.pixel_array[0,:,:]
print(arr.shape)
plt.imshow(arr,cmap=plt.cm.bone)
```
